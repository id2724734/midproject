```python
gajung2=train.groupby(["gu","age_group_rate"])["apt_age_group"].count()
gajung2 = pd.DataFrame(gajung2)

gajung2
```

```python
a = []
b = []
for j in train['gu'].unique():
    gajung2_yc=gajung2.loc[j]
    gajung2_yc=gajung2_yc.reset_index()
    # gajung_gn

    gagung2_sum=0
    sum2 = 0
    sum1=0

    for i in range(len(gajung2_yc["age_group_rate"])) : 


        sum1 = gajung2_yc["age_group_rate"][i] * gajung2_yc["apt_age_group"][i]
        gagung2_sum = gagung2_sum + sum1

        sum2 = sum2+ gajung2_yc["apt_age_group"][i]

    gagung2_avg_yc = gagung2_sum / sum2

    a.append(j)
    b.append(gagung2_avg_yc)
#     print(j, gagung2_avg_yc)
#     print('')
```
