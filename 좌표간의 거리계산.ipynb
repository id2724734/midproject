```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('data/real_train.csv', index_col = [0], encoding = 'euc-kr')
```

```python
train.head()
```

```python
gu_dis = pd.read_csv('data/seoul_gu_x_y.csv', index_col = [0], encoding = 'euc-kr')
gu_dis.loc[gu_dis['시군구명_한글'] == '강남구']
```

```python
from haversine import haversine
```

```python
gu_dis_x = list(gu_dis['위도'])
gu_dis_y = list(gu_dis['경도'])
```

```python
# for i in range(len(train_x)):
distance = []
for j in range(len(gu_dis_x)):
    start = (37.495985, 127.066409)
    goal = (gu_dis_x[j], gu_dis_y[j])
    distance.append(round(haversine(start, goal), 3))
distance
```

```python
len(distance)
```

```python
gu_dis['강남과의 거리'] = distance
gu_dis
```

```python
gu_kangnam = gu_dis.sort_values(by = ['강남과의 거리'])
gu_kangnam

gu_kangnam['강남까지'] = [i for i in range(len(gu_dis))]
gu_kangnam
```

```python
gu_kangnam['gu'] = gu_kangnam['시군구명_한글']
gu_kangnam['kangnam'] = gu_kangnam['강남까지']
del gu_kangnam['시군구코드']
del gu_kangnam['시군구명_한글']
del gu_kangnam['시군구명_영문']
del gu_kangnam['ESRI_PK']
del gu_kangnam['위도']
del gu_kangnam['강남까지']
del gu_kangnam['경도']

gu_kangnam
```

```python
del train['강남과의 거리']
del train['kangnam']

train.head()
```

```python
train = pd.merge(train,gu_kangnam,on=['gu'], how='outer')
train
```

```python
b = []
c = []
for i in train['gu'].unique():
    a = train.loc[train['gu'] == i]['transaction_real_price'].mean()
    b.append(a)
    
b
```

```python
train['gu_real_mean'] = 0
for i in range(len(train['gu'].unique())):
    train['gu_real_mean'] = np.where(train['gu'] == train['gu'].unique()[i], b[i], train['gu_real_mean'])
```
