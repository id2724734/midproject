```python
market = pd.concat([market_08, market_09, market_10, market_11, market_12, market_13, market_14, market_15, market_16, market_17])
market
```

```python
market['shop'] = market['대형마트'] + market['백화점'] + market['쇼핑센터']
```

```python
market = market.rename({'기간':'transaction_year', '자치구':'gu'}, axis = 'columns')
```

```python
train = pd.merge(train,market,on=['gu', 'transaction_year'], how='outer')
train.head()
```
