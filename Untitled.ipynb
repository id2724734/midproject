```python
train = pd.read_csv('data/train_train_train_train.csv', index_col = [0], encoding = 'euc-kr')

train
```

```python
market = pd.read_csv('data/market.csv' , index_col = [0], encoding = 'euc-kr')
market_2017 = market.loc[market['transaction_year'] == 2017]
```

```python
train = pd.merge(train,market_2017,on=['gu'], how='outer')
train.head()
```

```python
plt.rcParams.update({'font.size': 10})
plt.figure(figsize = (20, 7))
b = sns.barplot(data= train, x= 'gu', y= "transaction_real_price", hue='백화점')
plt.xlabel('구')
plt.ylabel('실거래가')
plt.title("구별 백화점 개수에 따른 평균 가격")
```

```python
df_river_re = train.sort_values(by = ['transaction_real_price'], axis = 0 , ascending = False)
df_river_re
```

```python
plt.rcParams.update({'font.size': 10})
plt.figure(figsize = (20, 7))
b = sns.barplot(data= df_river_re, x= 'gu', y= "transaction_real_price", hue='river_re')
plt.xlabel('동')
plt.ylabel('실거래가')
plt.title("구별 한강 및 재개발 유뮤 에 따른 평균 가격")
```

```python
train.corr()['transaction_real_price']
```

```python
df_corr = train[['gu','dong', 'exclusive_use_area', 'hanriver', 'transaction_real_price', 'exclusive_use_area_group', 'hakgun_rank', 'river_re', '백화점']]
```

```python
#상관관계 확인
k=df_corr.shape[1] #히트맵 변수 갯수
corrmat = df_corr.corr() #변수간의 상관관계
cols = corrmat.nlargest(k, 'transaction_real_price')['transaction_real_price'].index #price기준으로 제일 큰순서대로 20개를 뽑아냄
cm = np.corrcoef(df_corr[cols].values.T)
f, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(data = cm, annot=True, square=True, fmt = '.2f', linewidths=.5, cmap='Reds', 
            yticklabels = cols.values, xticklabels = cols.values)
```
