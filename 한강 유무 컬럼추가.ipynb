```python
len(han)
```

```python
train['dong'].unique()
```

```python
hanriver = []
for i in train['dong']:
    if i in han:
        hanriver.append(1)
    else:
        hanriver.append(0)
        
hanriver
```

```python
len(hanriver)
```

```python
train['hanriver'] = hanriver
train
```

```python
train['floor'] = np.where(train['floor'] < -1, -1, train['floor'])
```

```python
train['transaction_year_month'] = train['transaction_year_month'].astype(str)
train['transaction_year'] = train['transaction_year_month'].str[:4].astype(int)
train['transaction_month'] = train['transaction_year_month'].str[4:6].astype(int)
train['transaction_year_month'] = train['transaction_year_month'].astype(int)
```

```python
train.head()
```

```python
train['transaction_date'] = np.where(train['transaction_date'] == '1~10', 0, train['transaction_date'])
train['transaction_date'] = np.where(train['transaction_date'] == '11~20', 1, train['transaction_date'])
train['transaction_date'] = np.where(train['transaction_date'] == '21~28', 2, train['transaction_date'])
train['transaction_date'] = np.where(train['transaction_date'] == '21~29', 2, train['transaction_date'])
train['transaction_date'] = np.where(train['transaction_date'] == '21~30', 2, train['transaction_date'])
train['transaction_date'] = np.where(train['transaction_date'] == '21~31', 2, train['transaction_date'])
```
